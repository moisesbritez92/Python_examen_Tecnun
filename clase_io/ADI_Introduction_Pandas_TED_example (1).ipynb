### Which talks produce the most discussion online?

We can organize the DataFrame based on the number of comments and keep the ones at the end since they are sorted in ascending order


```python
ted.sort_values('comments', ascending=True).head(10)
```

The problem with this is that older videos may have more comments not because they provoke more discussion, but because they have been available for longer time.

To see which ones provoke more discussion, we can normalize the number of comments by the number of times each video has been viewed.

We calculate this new variable and add it to the DataFrame as follows:

```python
ted['comments_per_view'] = ted['comments'] / ted['views']
```

Now we can sort them but by `comments_per_view`

```python
ted.sort_values(by = 'comments_per_view', ascending=False)
```

We can calculate the number of views per comment to see how many views are needed on average to get a comment

```python
ted['views_per_comment'] = ted.views / ted.comments
```

We can sort the DataFrame according to this new variable to see which talks have the fewest 'views_per_comment.' These will be the ones that provoke more discussion.

We should get the same as before since they are inverse to each other.

```python
ted.sort_values('views_per_comment', ascending=False).head()
```

### Visualize the distribution of comments

```python
ted['comments'].plot(kind= 'hist')
```

We have drawn a time plot. The time plot doesn't make sense. On the x-axis, we have the index of the talk, and on the y-axis, the number of comments

Better a histogram.

```python
plt.plot(ted.index, ted.comments)
```

We need to adjust the number of bars to obtain more informative data. It can be seen that there are many talks with fewer than 1,000 comments.

```python
df_clean = ted[ted['comments']<1000]
sns.histplot(df_clean.comments, bins=100, kde=True)
```

How many talks are we missing?

```python
len(ted[ted['comments'] >= 1000])
```

We can do the same with the query() method

```python
# número de charlas con >= 1000 comentarios
ted.query('comments >= 1000').shape[0]

# o para ver las filas
#ted.query('comments >= 1000')
```

```python
# or with .loc
ted.loc[ted.comments < 1000, 'comments'].plot(kind='hist')
```

We can also do a boxplot

```python
sns.boxplot(data=df_clean, x= 'comments', orient='h')
```
