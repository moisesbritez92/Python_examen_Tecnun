### Which talks produce the most discussion online?

We can organize the DataFrame based on the number of comments and keep the ones at the end since they are sorted in ascending order


```python
ted.sort_values('comments', ascending=True).tail()
```

The problem with this is that older videos may have more comments not because they provoke more discussion, but because they have been available for longer time.

To see which ones provoke more discussion, we can normalize the number of comments by the number of times each video has been viewed.

We calculate this new variable and add it to the DataFrame as follows:

```python
ted['comments_per_view'] = ted.comments / ted.views
```

Now we can sort them but by `comments_per_view`

```python
ted.sort_values('comments_per_view').tail()
```

We can calculate the number of views per comment to see how many views are needed on average to get a comment

```python
ted['views_per_comment'] = ted.views / ted.comments
```

We can sort the DataFrame according to this new variable to see which talks have the fewest 'views_per_comment.' These will be the ones that provoke more discussion.

We should get the same as before since they are inverse to each other.

```python
ted.sort_values('views_per_comment').head()
```

### How many ratings has each talk received?

The goal is to sum the number of ratings (even if they are of different types) that each talk has received

```python
ted.ratings_list[0]
# What we are looking to do is create a function that sums the "count" of the
# ratings of a talk and use the apply function to do it for all the talks.
```

```python
# Let's create the function in pieces.
# First, we try to obtain the first dictionary from the list, that is, the one corresponding to the first rating.
def get_num_ratings(list_of_dicts):
    return list_of_dicts[0]
```

```python
# Check that it's correct
get_num_ratings(ted.ratings_list[0])
```

```python
# Now we modify the function so that it returns the count of the first rating
# instead of the complete dictionary
def get_num_ratings(list_of_dicts):
  return list_of_dicts[0]['count']
```

```python
# Check that it's correct
get_num_ratings(ted.ratings_list[0])
```

```python
# We modify the function so that it calculates the sum of all the counts
# for a single talk.
def get_num_ratings(list_of_dicts):
    num = 0
    for d in list_of_dicts:
        num = num + d['count']
    return num
```

```python
# Check that it's correct
get_num_ratings(ted.ratings_list[0])
```

```python
# Check with another talk
ted.ratings_list[1]
get_num_ratings(ted.ratings_list[1])
```

```python
# Finally, we apply the function to the entire talk and save it in a new column.
ted["num_ratings"] = ted.ratings_list.apply(get_num_ratings)
```

```python
# Check
ted.num_ratings.describe()

# There are no negative values and the obtained values look good
```

### What are the jobs of the speakers who give the funniest talks?

####  First step: Count the number of 'funny' ratings

```python
ted.ratings_list.head()

# We can see that "funny" is not always the first rating. But do all the talks have a "funny" rating?
```

```python
ted.ratings.str.contains('Funny').value_counts()

# By obtaining 2550 Trues, we conclude that all the talks have a "funny" rating.
# We are checking this in the original "ratings" column.
```

```python
# Create a function to count the "funny" ratings
def get_funny_ratings(list_of_dicts):
    for d in list_of_dicts:
        if d['name'] == 'Funny':
            return d['count']
```

```python
# Check that it works in a case where 'funny' is not the first rating
ted.ratings_list[3]
```

```python
get_funny_ratings(ted.ratings_list[3])
```

```python
# Apply the function to all talks with the method `apply`
ted['funny_ratings'] = ted.ratings_list.apply(get_funny_ratings)
ted.funny_ratings.head()
```

#### Second step: Calculate the ratio of 'funny' ratings to the total ratings for each talk

```python
ted['funny_rate'] = ted.funny_ratings / ted.num_ratings
```

```python
# Let's organize the talks based on the "funny_rate" and look at the occupations
# of the speakers from the talks with the highest funny_rate.
ted.sort_values('funny_rate').speaker_occupation.tail(20)
```

```python
ted.iloc[685]
```

```python
# Talks with lowest funny_rate.
ted.sort_values('funny_rate').speaker_occupation.head(20)
```

#### Third step: Analyze the average 'funny rate' for each job.

```python
ted.groupby('speaker_occupation').funny_rate.mean().sort_values().tail()

# In any case, it might happen that we have very few instances for each of the occupations.
```

```python
# We can check it as follows:
ted.speaker_occupation.describe()

# It is observed that out of 2,544 non-NaN cases, 1,458 are different.
# Moreover, the most frequent occupation appears only 45 times.
```

#### Fourth step: Focus only on those jobs that are well represented in the data

```python
# Let's start by counting how many times each occupation appears.
occupation_counts = ted.speaker_occupation.value_counts()
occupation_counts
```

```python
# Let's keep only those occupations that appear five or more times.
top_occupations = occupation_counts[occupation_counts >= 5].index
top_occupations
```

```python
# We create a new DataFrame where we only include the talks given by speakers
# with any of the occupations shown above.
ted_top_occupations = ted[ted.speaker_occupation.isin(top_occupations)]
ted_top_occupations.shape
```

#### Fifth step: Reanalyze the 'funny rate' for the new dataframe

```python
ted_top_occupations.groupby('speaker_occupation').funny_rate.mean().sort_values()
```
